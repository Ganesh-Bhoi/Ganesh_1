```python
# Q10. Write function which perform addition,substraction and multiplication?
def add(x,y):
    return x+y
def sub(x,y):
    return x-y
def mul(x,y):
    return x*y
select = input('''Select multiple operation +,-,*
                  Enter + Add
                  Enter - sub
                  Enter * mul''')
n1 = int(input('Enter a 1 number'))
n2 = int(input('Enter a 2 number'))
if select == '+':
    print(x,'+',y,'=',add(x,y))
elif select == '-':
    print(x,'-',y,'=',sub(x,y))
elif select == '*':
    print(x,'*',y,'=',mul(x,y))
else:
    print('Invalid operation')
```

```python
# Q5. which two columns are strongly corelated?
corr = df.corr()
venue = df.groupby(['City','Branch']).count()
venue = venue.iloc[:-1,:1].reset_index()
venue = venue.pivot('City','Branch')
venue.fillna(0,inplace=True)
sns.set_style('dark')
plt.figure(figsize=(17,10))
sns.heatmap(data=venue,annot=True,cmap='rocket')
plt.xlabel('City',size=15)
plt.ylabel('Branch',size=15)
plt.yticks(size=12)
plt.xticks(size=12)
plt.show()
```
